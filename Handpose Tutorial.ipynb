```python
import mediapipe as mp
import cv2
import numpy as np
import uuid
import os
```

```python
mp_drawing = mp.solutions.drawing_utils#series of landmarks red dots
mp_hands = mp.solutions.hands #everything ab hnd
```

# 2. Draw Hands
<img src=https://i.imgur.com/qpRACer.png />

```python
cap = cv2.VideoCapture(0)#vedio capture devide 0 on windows

with mp_hands.Hands(min_detection_confidence=0.8, min_tracking_confidence=0.5) as hands: 
    while cap.isOpened():
        ret, frame = cap.read()#eacg frame from web cam, & return meeting
        
        # BGR 2 RGB
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        # Flip on horizontal
        image = cv2.flip(image, 1)
        
        # Set flag
        image.flags.writeable = False
        
        # Detections
        results = hands.process(image)
        
        # Set flag to true
        image.flags.writeable = True
        
        # RGB 2 BGR
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        # Detections
        print(results)
        
        # Rendering results
        if results.multi_hand_landmarks:
            for num, hand in enumerate(results.multi_hand_landmarks):
                mp_drawing.draw_landmarks(image, hand, mp_hands.HAND_CONNECTIONS, 
                                        mp_drawing.DrawingSpec(color=(121, 22, 76), thickness=2, circle_radius=4),
                                        mp_drawing.DrawingSpec(color=(250, 44, 250), thickness=2, circle_radius=2),
                                         )
            
        
        cv2.imshow('Hand Tracking', image)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```

# 3. Output Images 

```python
os.mkdir('Output Images')
```

```python
cap = cv2.VideoCapture(0)
#initiating the mediapipe hands model #80% for detection
with mp_hands.Hands(min_detection_confidence=0.8, min_tracking_confidence=0.5) as hands: 
    while cap.isOpened():                               ## if we want for more than 2 hands we can put max_num_hands
        ret, frame = cap.read()
        
        # BGR 2 RGB #recoloring the frame bcz by default we get frame as bgr format
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        # Flip on horizontal
        image = cv2.flip(image, 1)
        
        # Set flag
        image.flags.writeable = False
        
        # Detections
        results = hands.process(image)
        
        # Set flag to true
        image.flags.writeable = True #in order to draw on these images
        
        
        # RGB 2 BGR
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        # Detections
        print(results)
        
        # Rendering results
        if results.multi_hand_landmarks:#got any landmark
            for num, hand in enumerate(results.multi_hand_landmarks):
                mp_drawing.draw_landmarks(image, hand, mp_hands.HAND_CONNECTIONS, #to change color
                                        mp_drawing.DrawingSpec(color=(121, 22, 76), thickness=2, circle_radius=4),#landmark
                                        mp_drawing.DrawingSpec(color=(150, 44, 250), thickness=2, circle_radius=2),#lines
                                         )
            
        # Save our image    #to save our images
        cv2.imwrite(
            os.path.join(
                'Output Images',
                '{}.jpg'.format(uuid.uuid1())), image) #naming our image we use uuid creates unique id as string which
                                                       # is used as name of image
        cv2.imshow('Hand Tracking', image)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```
